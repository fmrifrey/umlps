```python
from vis3d import tim
import torch
import numpy as np
import h5py
```

```python
with h5py.File('/home/djfrey/data/acr_testing_20250407/recon.h5', 'r') as h5_file:
    img = torch.tensor(h5_file['sol/real'][:] + 1j * h5_file['sol/imag'][:]) # kspace data
img = img.permute(0,1,4,3,2)
```
